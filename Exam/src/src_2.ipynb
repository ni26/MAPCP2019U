```python
# Now we try to find the c and lambda coefficient for our model

import numpy as np
def getCellCount(time,lamda,coef):
    from numpy import exp
    return 100000.0 * exp( lamda * ( 1.0 - exp( -coef * time ) ) )

GetCellCount = np.vectorize(getCellCount)

def getSumSquaredDiff(ParamVec):
    import numpy as np
    TimeVec = np.double([ 0.0, 10.0, 12.0, 14.0, 16.0, 18.0, 20.0, 22.0 ])
    CellVec = np.double([ 100000,15353211.952172322,45042595.52870743,74487574.87027894,86011263.45784552,
                         102072672.96304679,121122958.37704065,146417628.48747736])
    ExpectedCellCount = GetCellCount( TimeVec, ParamVec[0], ParamVec[1] )
    return np.sum( ExpectedCellCount - CellVec ) ** 2

from scipy.optimize import fmin
BestParam = fmin(getSumSquaredDiff, x0 = np.double([1,0]))
```

```python
# So we happily have the lamda and coef for our model. 
# So we can happily plog these values in to see how we can plot it

fig = plt.figure()
ax = plt.axes()
plt.xlabel('Time [days]')
plt.ylabel('Tumor Cell Count')
plt.title("Gompertzian Fit to Rat's Brain Tumor Growth")
plt.legend(["Legend"])

x = np.linspace(0, 10, 1000)
ax.plot(x, 100000.0 * np.exp( 6.98649434 * ( 1.0 - np.exp( -0.2264074 * x ) ) ));
```
